```python
import pandas as pd
import re
```

```python
# pd = pd.read_csv('to_download.txt', sep=' 	', header=None)
file_path = 'data/to_download.txt'

with open(file_path, 'r') as file:
    lines = file.readlines()

# pattern = re.compile(r'(\d{4} \w+ \(\w+-\w+\)-\w+\.csv)\s+(\w+ \d{1,2}[a-z]{2} \d{4}, \d{2}:\d{2}:\d{2} [ap]m)\s+([\d.]+ \wB)\s+(CSV file)')
pattern = re.compile(r'(.*?\.csv)\s+(\w+ \d{1,2}[a-z]{2} \d{4}, \d{2}:\d{2}:\d{2} [ap]m)\s+([\d.]+ \wB)\s+(CSV file)')

data = []
for line in lines:
    match = pattern.match(line)
    if match:
        data.append(match.groups())

# Create a DataFrame
df = pd.DataFrame(data, columns=['File Name', 'Date', 'Size', 'Type'])

df['File Name No Space'] = df['File Name'].str.replace(' ', '%20')
```
