```python
import pandas as pd
```

```python
journey_df = pd.read_csv('data/Journey_Data/394JourneyDataExtract15Apr2024-30Apr2024.csv')
journey_df.head()
```

```python
file_path = 'data/docking_station_locations.txt'
header = ['Station.Id','StationName','longitude','latitude','Easting','Northing']

docking_df = pd.read_csv(file_path, sep=',', encoding='utf-8', header=0, names=header)
print(docking_df.head())
```

```python
df_temp = docking_df[['StationName', 'longitude', 'latitude']]
# start
df_temp.columns = ['Start station', 'start_longitude', 'start_latitude']
journey_loc_df = journey_df.merge(df_temp, how='left', on='Start station')
# end
df_temp.columns = ['End station', 'end_longitude', 'end_latitude']
journey_loc_df = journey_loc_df.merge(df_temp, how='left', on='End station')

journey_loc_df.head()

journey_loc_df.to_pickle('pickle/journey_loc_df')
```
